```python
import random
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
data = pd.read_csv('/content/Tata-steel.csv')
X = data[['Open Price','High Price','Low Price']]
y = data[['Close Price']]
x_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle = False)
```

```python
y_test = y_test.reset_index()[["Close Price"]]
y_test
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.metrics import mean_squared_error



def BuildModel():
 model = Sequential()
 model.add(Dense(30, input_dim=3,activation='relu'))  
 model.add(Dense(1,activation='linear'))
 model.compile(loss="mean_squared_error", optimizer="adam")   
 return model

BuildModel().summary()
regressor = KerasRegressor(build_fn=BuildModel,nb_epoch=400,batch_size=8)
regressor.fit(x_train,y_train) 

y_pred = regressor.predict(X_test)
mse_krr = mean_squared_error(y_test, y_pred)
print(mse_krr)


plt.figure(figsize=(30,10), dpi=80)
plt.rcParams.update({'font.size': 18})
plt.plot(y_test, 'bo-', label="Actual close prices")
plt.plot(y_pred, 'ro-', label="Predicted close prices")
plt.legend(prop={'size': 20})
plt.show()
```
